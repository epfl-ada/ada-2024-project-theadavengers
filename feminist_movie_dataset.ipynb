```python
import requests
from bs4 import BeautifulSoup

url = "https://letterboxd.com/brunaleo/list/essential-feminist-films-worldwide/detail/page/2/"
# Fetch the HTML content
response = requests.get(url)
soup = BeautifulSoup(response.content, "html.parser")
# Extract the list items
items = soup.select("h2.headline-2 a")

url2 = "https://letterboxd.com/brunaleo/list/essential-feminist-films-worldwide/detail/"
# Fetch the HTML content
response = requests.get(url2)
soup = BeautifulSoup(response.content, "html.parser")
# Extract the list items
for item in  soup.select("h2.headline-2 a"):
    items.append(item)

url3 = "https://letterboxd.com/brunaleo/list/essential-feminist-films-worldwide/detail/page/3/"
# Fetch the HTML content
response = requests.get(url3)
soup = BeautifulSoup(response.content, "html.parser")
# Extract the list items
for item in  soup.select("h2.headline-2 a"):
    items.append(item)

url4 = "https://letterboxd.com/brunaleo/list/essential-feminist-films-worldwide/detail/page/4/"
# Fetch the HTML content
response = requests.get(url4)
soup = BeautifulSoup(response.content, "html.parser")
# Extract the list items
for item in  soup.select("h2.headline-2 a"):
    items.append(item)

url5 = "https://letterboxd.com/brunaleo/list/essential-feminist-films-worldwide/detail/page/5/"
# Fetch the HTML content
response = requests.get(url5)
soup = BeautifulSoup(response.content, "html.parser")
# Extract the list items
for item in  soup.select("h2.headline-2 a"):
    items.append(item)

url6 = "https://letterboxd.com/brunaleo/list/essential-feminist-films-worldwide/detail/page/6/"
# Fetch the HTML content
response = requests.get(url6)
soup = BeautifulSoup(response.content, "html.parser")
# Extract the list items
for item in  soup.select("h2.headline-2 a"):
    items.append(item)
```

```python
print(len(items))
```

```python
for item in items:
    print(item.get_text(strip=True))
```

```python
print(items[0].get_text(strip=True))
print(items[1].get_text(strip=True))
```

```python
movie_titles = []
movie_years = []
i = 0
for item in items:
    if i%2 == 0:
        title = item.get_text(strip=True)
        movie_titles.append(title)
    else :
        year = item.get_text(strip=True)
        movie_years.append(int(year))
    i+=1
```

```python
print(len(movie_years))
print(len(movie_titles))
```

```python
import pandas as pd

dict = {'Movie name': movie_titles, 'Year': movie_years} 
   
feminist_movies_df = pd.DataFrame(dict)
```
